```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from tabulate import tabulate
# services
```

```python
options = Options()
# options.add_argument("--headless")

driver = webdriver.Chrome(options=options)
driver.get('https://www.psx.com.pk/market-summary/')
```

```python
tables = driver.find_elements(By.CLASS_NAME, 'table-responsive')
tables
```

```python
for table in tables:
    header = table.find_element(By.TAG_NAME,'tr').text
    table_body = table.find_element(By.TAG_NAME,'tbody').text
    print(header, table_body)                              
```

```python
for table in tables:
    # Get all rows
    rows = table.find_elements(By.TAG_NAME, "tr")
    
    # Extract headers
    headers = [th.text for th in rows[0].find_elements(By.TAG_NAME, "th")]

    # Extract all rows after header
    data = []
    for row in rows[1:]:
        cells = row.find_elements(By.TAG_NAME, "td")
        data.append([cell.text for cell in cells])
    # Print each table nicely
    print("\n" + tabulate(data, headers=headers, tablefmt="Grids"))
    
```

```python
import csv
from selenium.webdriver.common.by import By

data = []
current_category = ""

for table in tables:  # Assuming you already located the correct tables
    rows = table.find_elements(By.TAG_NAME, 'tr')

    for row in rows:
        # Get header or category rows
        headers = row.find_elements(By.TAG_NAME, 'th')
        if headers:
            header_text = headers[0].text.strip()
            if header_text and header_text != "SCRIP":
                current_category = header_text
            continue

        cols = row.find_elements(By.TAG_NAME, 'td')
        if len(cols) == 8:
            scrip = cols[0].text.strip()
            ldcp = cols[1].text.strip()
            open_ = cols[2].text.strip()
            high = cols[3].text.strip()
            low = cols[4].text.strip()
            current = cols[5].text.strip()
            change = cols[6].text.strip()
            volume = cols[7].text.strip()

            data.append([
                current_category,
                scrip,
                ldcp,
                open_,
                high,
                low,
                current,
                change,
                volume
            ])

# Save to CSV
with open('market_data.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow([
        'CATEGORY', 'SCRIP', 'LDCP', 'OPEN', 'HIGH', 'LOW', 'CURRENT', 'CHANGE', 'VOLUME'
    ])
    writer.writerows(data)

print(" Data saved to market_data.csv")
```
